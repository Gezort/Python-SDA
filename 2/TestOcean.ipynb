```python
from B import Ocean
from matplotlib import pyplot as plt
from numpy.random import randint
%matplotlib inline
```

```python
plt.figure(figsize=(20, 10))
for m in range(100,105):
    for n in range(100,105):
        num_predators = randint(1, high=n * m // 3)
        num_victims = randint(num_predators, high=2 * m * n // 3)
        num_barriers = randint(0, n * m // 100)
        t1 = randint(3,5 + 15)
        t2 = randint(3,5 + 10)
        b = list(zip(randint(0, high=n, size=num_barriers), randint(0, high=m, size=num_barriers)))
        v = list(zip(randint(0, high=n, size=num_victims), randint(0, high=m, size=num_victims)))
        p = list(zip(randint(0, high=n, size=num_predators), randint(0, high=m, size=num_predators)))
        ocean = Ocean(width=m, height=n, starvation_time=t1, birth_time=t2, 
                      victims=v, predators=p, barriers=b)
        result = ocean.run_simulation(100, console_simulation=False)
        time = len(result)
        plt.subplot(5, 5, 5 * (m - 100) + (n - 99))
        plt.plot(range(0, time), [p[0] for p in result], 'g')
        plt.plot(range(0, time), [p[1] for p in result], 'r')
        l1 = plt.xlabel('time')
        l2 = plt.ylabel('number of creatures')
plt.legend(['victims', 'predators'])
plt.show()
```
